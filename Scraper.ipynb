```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
# get the response in the form of html
engWikiUrl="https://en.wikipedia.org/wiki/List_of_Sri_Lankan_monarchs"
sinWikiUrl="https://si.wikipedia.org/wiki/%E0%B6%BD%E0%B6%9A%E0%B7%8A%E0%B6%AF%E0%B7%92%E0%B7%80_%E0%B6%BB%E0%B7%8F%E0%B6%A2%E0%B7%8F%E0%B7%80%E0%B6%BD%E0%B7%92%E0%B6%BA"
table_class="wikitable sortable jquery-tablesorter"
engResponse=requests.get(engWikiUrl)
sinResponse=requests.get(sinWikiUrl)
print(engResponse.status_code,sinResponse.status_code)
```

```python
soup = BeautifulSoup(engResponse.text, 'html.parser')
engData = pd.DataFrame()
for indiatable in soup.find_all('table',{'class':"wikitable"}):
    df=pd.read_html(str(indiatable))
    df=pd.DataFrame(df[0])
    #if 'පූර්වප්‍රාප්තික(යන්) සමග සම්බන්ධතාවය' in df.columns:
    engData = pd.concat([engData,df])
    #print(df.columns.equals('රුව', 'නම', 'උපත', 'මරණය', 'රාජ්‍ය සමයේ ආරම්භය', 'රාජ්‍ය සමයේ අවසානය','පූර්වප්‍රාප්තික(යන්) සමග සම්බන්ධතාවය')
    indiatable.findAll('a')
engData.head()
```

```python
soup = BeautifulSoup(sinResponse.text, 'html.parser')
sinData = pd.DataFrame()
for indiatable in soup.find_all('table',{'class':"wikitable"}):
    df=pd.read_html(str(indiatable))
    df=pd.DataFrame(df[0])
    sinData = pd.concat([sinData,df])
    indiatable.findAll('a')
sinData.head()
```
